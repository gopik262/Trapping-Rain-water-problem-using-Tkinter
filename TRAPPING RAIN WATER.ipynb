```python
from tkinter import *

class Window(Frame):

    def __init__(self, master=None):
            Frame.__init__(self, master)        
            self.master = master

            self.pack(fill=BOTH, expand=1)

           
            lbl=Label(self, text="Enter the size of Block give space:", fg='green', font=("Helvetica", 16))
            lbl.place(x=10, y=50)

           
            self.blockSize=Entry(self, bd=5)
            self.blockSize.place(x=120, y=100)

           
            btn=Button(self, text="Calculate", command = self.findWater)
            btn.place(x=150, y=150)

            self.lbl1=Label(self, font=("Helvetica", 12))
            self.lbl1.place(x=10, y=200)

   
    def findWater(self):
    
        arr = list(map(int, self.blockSize.get().split()))
        n = len(arr)
 
        result = 0
        left_max = 0
        right_max = 0
        lo = 0
        hi = n-1
        while(lo <= hi):
            if(arr[lo] < arr[hi]):
            
                if(arr[lo] > left_max):
    
                
                    left_max = arr[lo]
                else:
    
                    
                    result += left_max - arr[lo]
                lo+= 1
            
            else:
            
                if(arr[hi] > right_max):
                    
                    right_max = arr[hi]
                else:
                    result += right_max - arr[hi]
                hi-= 1
            
        
        self.lbl1['text'] = "Maximum water that can be accumulated is " + str(result)
        

root = Tk()
app = Window(root)
root.wm_title("Trapping Rain Water")
root.geometry("500x500")
root.mainloop()
```
